```python
#sasuke uchiha
from PIL import Image
import numpy as np
import random


accessoriesHairs = ["Empty","Empty","Empty" ,"Empty","Empty","Empty","Empty","Empty","Empty","Empty","Empty","Empty","Empty","Empty"]
accessoriesEyes = ["Empty", "Empty","Empty","Empty","Empty","Empty","Empty","Empty"]
accessoriesSowrd = ["Empty","Empty","Empty","Empty","Empty","Empty","Empty","Empty","Empty"]
#accessoriesHairs  =  ["hair.png","hair2.png","Empty","Empty", "Empty","Empty","Empty","Empty","Empty","Empty","Empty"] 
#accessoriesEyes  =  ["eyes.png","Empty","Empty", "Empty","Empty","Empty","Empty","Empty","Empty","Empty"]
#accessoriesSword =  ["Empty","Empty", "Empty","Empty","Empty","Empty","Empty","Empty","Empty"]

basePath = r"C:/Users/adhrit/OneDrive/Desktop/fun/nft/mytry/input/"
generatedImagePath = r"C:/Users/adhrit/OneDrive/Desktop/fun/nft/mytry/results/"

def alteredImageColor(im, originalColour, newColour):
    data = np.array(im)
    r1,g1,b1 = originalColour[0], originalColour[1],originalColour[2]
    r2,g2,b2 = newColour[0], newColour[1],newColour[2]
    
    red, green, blue = data[:,:,0],data[:,:,1],data[:,:,2]
    mask = (red == r1) & (green == g1) & (blue == b1)
    data[:,:,:3][mask] = [r2, g2,b2]
    
    newImage = Image.fromarray(data)
    return newImage

def getRandomColourArray():
    return [random.randint(0,255), random.randint(0,255),random.randint(0,255)]

def getBaseImage():
    baseImage = Image.open(basePath + "alien2.jpg")
    baseImage = alteredImageColor(baseImage, [255, 0, 155], getRandomColourArray())
    return baseImage

def getHairImage():
    hairColours = Image.open(basePath + "alien2.jpg" ).convert("RGBA")
    hairColours = alteredImageColor(hairColours,[0,162,232], getRandomColourArray())
    hairColours = alteredImageColor(hairColours,[255, 255, 255], getRandomColourArray())
    hairColours = alteredImageColor(hairColours, [153,217,234], getRandomColourArray())
    return hairColours


def getFaceImage():
    faceColour = Image.open(basePath + "alien3.jpg" )
    faceColour = alteredImageColor(faceColour , [237, 28,36], getRandomColourArray())
    return faceColour


def getOutlineImage():
    outline = Image.open(basePath + "Background.jpg").convert("RGBA")
    return outline

def combineImages(baseImage, onTopImage):
    baseImage.paste(onTopImage,(0,0), mask = onTopImage)
    return baseImage

def applyAccessory(baseImage, accessoryList):
    accessory = accessoryList[random.randint(0,len(accessoryList) - 1 )]
    if accessory != "Empty":
        accessoryImage = Image.open(basePath + accessory).convert("RGBA")
        baseImage.paste(accessoryImage, (0,0), mask = accessoryImage)
        
def saveHighResImage(baseImage, imageID):
    baseImage =baseImage.resize((500,500),Image.NEAREST)
    imageLocation = generatedImagePath + "/Anime" + str(imageID) + ".png"
    baseImage.save(imageLocation)
  
def makeAnime(labelNumber):
    baseImage = getBaseImage()
    outlineImage = getOutlineImage()
    #faceImage = getFaceImage()
    hairImage = getHairImage()
    
    combineImages(baseImage, outlineImage)
    combineImages(baseImage, hairImage)
    #combineImages(baseImage, faceImage)
    
    applyAccessory(baseImage, accessoriesHairs)
    applyAccessory(baseImage, accessoriesEyes)
    #applyAccessory(baseImage, accessoriesSowrd)
            
    saveHighResImage(baseImage, labelNumber)

    
def makeMultiAnime(count):
    for i in range(count):
        makeAnime(i)

makeMultiAnime(50)
```
